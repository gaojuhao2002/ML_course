```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python

plt.figure(figsize=(15,5),dpi=100)
matplotlib.rcParams['font.family']='STSong'
df=pd.read_csv('dataForExercise.csv',header=None,names=['x','y'])
df_sampled=df[::50]
plt.plot(df['x'],df['y'])
plt.scatter(df_sampled['x'],df_sampled['y'],c='red')
plt.legend(['',''])
plt.title('dataForExercise.csv')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
